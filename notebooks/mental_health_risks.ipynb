```python
num_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(20,20))
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")

plt.title("Mental Health Correlation")

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)              

plt.tight_layout()
plt.show()
```

```python
corr = num_df.corr()

target_corr = corr['mental_health_risk'].drop('mental_health_risk')
```

```python
top_10 = target_corr.abs().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))

top_10.sort_values().plot(kind='barh')

plt.title("Top 10 Predictors of Mental Health Risk")
plt.xlabel("Correlation Strength (absolute value)")
plt.ylabel("Features")

plt.show()
```

```python
top_10_signed = target_corr.loc[top_10.index]

plt.figure(figsize=(10,6))

top_10_signed.sort_values().plot(kind='barh')

plt.title("Top 10 Predictors of Mental Health Risk (Directional)")
plt.xlabel("Correlation (positive or negative)")
plt.ylabel("Features")

plt.axvline(0, color='black', linewidth=1)

plt.show()
```

## Key Predictors of Mental Health Risk

The strongest predictors of mental health risk include stress level, depression score, sleep duration, and screen time.

Higher stress and depression scores show a positive correlation with mental health risk, while factors like sleep and social support show negative correlations, suggesting protective effects.

## Key Findings

- Higher stress levels are strongly associated with higher mental health risk.
- Lower sleep hours correlate with increased risk levels.
- Screen time appears positively associated with risk.
- Social support shows a negative relationship with risk (protective factor).
- Demographic variables (gender, education) show weaker correlation with risk.

## Conclusion

This analysis highlights lifestyle and behavioral factors as stronger indicators of mental health risk compared to demographic variables. Stress, sleep, and screen time appear to be the most influential features in predicting risk levels.
