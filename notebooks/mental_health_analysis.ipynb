```python
categorical_cols = [
    'Sex',
    'Race',
    'Education_Level',
    'Employment_Status',
    'Housing_Status',
    'Insurance_Status',
    'Access_to_Healthcare',
    'Primary_Care_Access',
    'Substance_Use',
    'Depression_Risk',
    'Anxiety_Risk',
    'Suicide_Risk',
    'Neighborhood'
]

for col in categorical_cols:
    print(f"\n{'='*40}")
    print(f"{col.upper()} DISTRIBUTION")
    print(f"{'='*40}")
    
    counts = df[col].value_counts()
    percentages = df[col].value_counts(normalize=True) * 100
    
    summary = pd.DataFrame({
        'Count': counts,
        'Percentage (%)': percentages.round(2)
    })
    
    display(summary)
    
    plt.figure()
    counts.plot(kind='bar')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
plt.figure()
plt.scatter(df['Income'], df['Mental_Health_Risk_Score'])
plt.title('Income vs Mental Health Risk')
plt.xlabel('Income')
plt.ylabel('Mental Health Risk Score')
plt.show()
```

```python
plt.figure()
plt.scatter(df['Substance_Use'], df['Mental_Health_Risk_Score'])
plt.title('Substance Use vs Mental Health Risk')
plt.xlabel('Substance Use (0=None, 1=Occasional, 2=Frequent)')
plt.ylabel('Risk Score')
plt.show()
```

```python
plt.figure()
plt.scatter(df['Access_to_Healthcare'], df['Mental_Health_Risk_Score'])
plt.title('Healthcare Access vs Mental Health Risk')
plt.xlabel('Access (0=None, 1=Limited, 2=Good)')
plt.ylabel('Risk Score')
plt.show()
```

```python
plt.figure(figsize=(10,8))
corr = df.corr(numeric_only=True)

plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title('Correlation Heatmap')
plt.show()
```

```python
df.groupby('Employment_Status')['Mental_Health_Risk_Score'].mean().plot(kind='bar')
plt.title('Employment Status vs Mental Health Risk')
plt.ylabel('Average Risk Score')
plt.show()
```

```python
pd.crosstab(df['Substance_Use'], df['Suicide_Risk']).plot(kind='bar')
plt.title('Substance Use vs Suicide Risk')
plt.ylabel('Count')
plt.show()
```

```python
df.groupby('Neighborhood')['Mental_Health_Risk_Score'].mean().plot(kind='bar')
plt.title('Neighborhood vs Mental Health Risk')
plt.ylabel('Average Risk Score')
plt.show()
```

```python
df.groupby(['Employment_Status', 'Substance_Use'])['Mental_Health_Risk_Score'] \
  .mean().unstack().plot(kind='bar')

plt.title('Employment + Substance Use Impact on Mental Health')
plt.ylabel('Average Risk Score')
plt.show()
```

```python
plt.figure()
df['Mental_Health_Risk_Score'].hist()
plt.title('Mental Health Risk Score Distribution')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

```python
pivot = df.pivot_table(
    values='Substance_Use',
    index='Neighborhood',
    columns='Race',
    aggfunc='mean'
)

plt.figure()

plt.imshow(pivot)
plt.colorbar()

plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=90)
plt.yticks(range(len(pivot.index)), pivot.index)

# 🔥 ADD ANNOTATIONS
for i in range(len(pivot.index)):
    for j in range(len(pivot.columns)):
        value = pivot.iloc[i, j]
        if not pd.isna(value):  # avoid NaN labels
            plt.text(j, i, f"{value:.2f}", ha='center', va='center')

plt.title('Average Substance Use by Neighborhood and Race')
plt.show()
```
